```julia
using MathOptInterface, JuMP, Cbc
const MOI = MathOptInterface
using Cbc


m = Model(Cbc.Optimizer)

c = [ 1; 4; 9]
A = [1  1  0;
     1  0  1;
     0 -1  1]

b = [5; 10; 7]
lb = [-Inf;-1;-Inf]
ub = [4;Inf;1]
index_x = 1:3
index_constraints = 1:2
index_constraints2 = 3

                       
@variable(m, x[i in index_x], lower_bound = lb[i], upper_bound=ub[i])
var=all_variables(m)

@objective(m, Min, sum( c[i]*var[i] for i in index_x) )

@constraint(m, constraint1,
               sum( A[1,i]*var[i] for i in index_x ) <= b[1] )
@constraint(m, constraint2,
               sum( A[2,i]*var[i] for i in index_x ) >= b[2] )
@constraint(m, constraint3,
               sum( A[3,i]*var[i] for i in index_x ) == b[3] )

print(m)
```

```julia
using MathOptInterface, JuMP, Cbc
const MOI = MathOptInterface
using Cbc


m = Model(Cbc.Optimizer)

c = [ 1; 4; 9; -5; 5; 3; 7; 8; 9; 7; -8]

A =  [0  1  0   0  1  2  3  0  0  4  -1;
      1  0  0  -1  1  2  0  4  1  -2  5;
      0 -1  1   1  0  2  1 -5  1  -3  0]

b = [ 0; 0; 0; 100; 1; 2; 1; 50;  2;  10; 5]
lb= [-1;-1;4;10;0;4;5;1;0;-Inf;-Inf]
ub = [Inf;Inf;Inf;Inf;Inf;Inf;Inf;Inf;Inf;Inf;Inf]
index_x = 1:11
index_constraints = 1:3


@variable(m, x[i in index_x], lower_bound = lb[i], upper_bound=ub[i])
var=all_variables(m)
@objective(m, Max, sum( c[i]*var[i] for i in index_x) )
@constraint(m, constraint[j in index_constraints],
               sum( A[j,i]*var[i] for i in index_x ) >= b[j] )
                       
print(m)

JuMP.optimize!(m)
```

```julia
println("Optimal solutions:")
println("x1 = ", JuMP.value(var[1]))
println("x2 = ", JuMP.value(var[2]))
println("x3 = ", JuMP.value(var[3]))

println("The objective vablue = ", JuMP.objective_value(m))
```
